# Average sale month on month

```python
df['Order_date']=pd.to_datetime(df['Order_date'])

```

```python
df[['Order_date','Sale_price']].groupby(by=['Order_date','Sale_price']).sum()
```

```python
a=df[['Sale_price']].groupby(by=[df.Order_date.dt.month]).mean()
a.rename(columns={'Sale_price':'Avg_Sale_price'}, inplace=True)
a
```

# Average order MoM

```python
b=df[['Order_id']].groupby(by=[df.Order_date.dt.month]).count()
b.rename(columns={'Order_id':'total_orders'},inplace=True)
b
```

```python
df['Month']=df.Order_date.dt.month
df
```


# Total order X gender MoM

```python
e=df.pivot_table(index=['Month'],columns=['Gender'],values=['Order_id'],aggfunc='count',margins=True,margins_name='total')
e.rename(columns={'Order_id':'total_orders'},inplace=True)
e
```

# Average sale X Product category MoM

```python
c=df.pivot_table(index=['Month'],columns=['Product_category'],values=['Sale_price'])
c.rename(columns={'Sale_price':'Avg_sale_price'},inplace=True)
c
```

# Total sale X Product category MoM

```python
d=df.pivot_table(index=['Month'],columns=['Product_category'],values=['Sale_price'],aggfunc='sum',margins=True,margins_name='Grand_total')
d.rename(columns={'Sale_price':'Total_sale'},inplace=True)
d
```

# Order rating X Product category MoM/Over the time

```python
f=df.pivot_table(index=['Order_rating'],columns=['Product_category'],values='User_id',aggfunc='count',margins=True,margins_name='Grand_total')
f
```

# Calculate order_status MoM

```python
g=df.pivot_table(index=['Month'],columns=['Order_status'],values='User_id',aggfunc='count',margins=True,margins_name='Grand_total')
g
```

# Product portfolio OC

```python
order_contribution=df[['Product_portfolio','User_id']].groupby(by=['Product_portfolio']).count()
order_contribution.rename(columns={'User_id':'Total_orders'},inplace=True)
order_contribution
```

```python
order_contribution['Order_contribution(in%)']=order_contribution.Total_orders*100/order_contribution.Total_orders.sum()
order_contribution
```

# New Users MoM

```python
monthly_users=df[['Month','User_id']].groupby(by=['Month']).nunique()
monthly_users['New_users']=monthly_users['User_id'].diff()
monthly_users.rename(columns={'User_id':'Total_users'},inplace=True)
monthly_users
```

# Repeat Users MoM

```python
monthly_repeat_users=df[['Month','User_id']].groupby(by=['Month']).nunique()
monthly_repeat_users.rename(columns={'User_id':'Total_users'},inplace=True)
monthly_repeat_users['Repeat_users']=monthly_repeat_users['Total_users']-monthly_repeat_users['Total_users'].diff()
monthly_repeat_users
```
